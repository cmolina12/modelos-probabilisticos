Los clientes llegan a una peluquería con una tasa media de 5 clientes/hora y los tiempos de llegada se distribuyen exponencialmente. Siempre hay un peluquero presente en la peluquería en cualquier momento y hay 4 sillas para los clientes que llegan cuando el peluquero está ocupado. La ley de prevención de incendios limita el número total de clientes en el salón, en cualquier momento, a un máximo de 5. Los clientes que llegan cuando el salón está lleno no pueden entrar en él. El tiempo de servicio se distribuye exponencialmente con una media que depende del número de clientes. Determina: 

- Servidor: 1
- Capacidad: 5

```python
from math import factorial
from math import prod
```

```python
# Funcion para en caso de M/M/1/K con tasa mu variable segun el numero de clientes en el sistema, sacar la probabilidad de estar vacio

lmbda = 5  # clientes/hora

mu = {
    1: 60 / 9,
    2: 60 / 10,
    3: 60 / 10,
    4: 60 / 13,
    5: 60 / 20,
}  # diccionario de mu, llave siendo numero de clientes en el sistema y valor siendo la tasa media de servicio

# Sacar el diccinario de cada mu multiplicado por todos sus mu_s anteriores

mu_dict = {n: prod(mu[i] for i in range(1, n + 1)) for n in mu}

print(mu_dict)

k = 5  # numero maximo de clientes en el sistema

s = 1  # 1 peluquero

# 1. Coeficientes de equilibrio para cada p_i

coeficientes = {n: (lmbda**n / mu_dict[n]) for n in mu}

print(coeficientes)

# 2. p_0

p_0 = 1 / (1 + sum(coeficientes[n] for n in mu))

print(f"p_0: {p_0}")

# 3. p_n

p_n = {n: p_0 * coeficientes[n] for n in mu}

for i in p_n:
    print(f"p_{i}: {p_n[i]}")
```

## a) Numero medio de clientes en espera

```python
L = sum(n * p_n[n] for n in mu)
print(f"L: {L}")

Lq = L - (1-p_0)

print(f"Lq: {Lq}")
```

## b) Tiempo medio de espera

```python
lmbda_efectivo = lmbda*(1-p_n[5])
print(f"lmbda_efectivo: {lmbda_efectivo}")

W_q = Lq/lmbda_efectivo
print(f"W_q: {W_q} horas")
print(f"W_q: {W_q*60} minutos")
```
